```python
# Crear en Repl.it un programa que evalue las compruebas And y Or y Not


valor1 = True
valor2 = False

print(f"Valor 1: {valor1}")
print(f"Valor 2: {valor2}\n")

print(f"Operación: {valor1} AND {valor2} = {valor1 and valor2}")
print(f"Operación: True AND True = {True and True}")
print(f"Operación: False AND False = {False and False}\n")

print(f"Operación: {valor1} OR {valor2} = {valor1 or valor2}")
print(f"Operación: True OR True = {True or True}")
print(f"Operación: False OR False = {False or False}\n")

print(f"Operación: NOT {valor1} = {not valor1}")
print(f"Operación: NOT {valor2} = {not valor2}\n")

```
